```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("weather.csv", parse_dates=["day"])
df.head()
type(df.day[0])
df.head()
# type(df.day[0])
df.set_index("day", inplace=True)
df.head()
```

```python
df1 = df.fillna(method='ffill', axis='columns')
# print(df1)
df2 = df.fillna({
    'temperature':30,
    'windspeed':8,
    'event':"No Event"
})
df2
# specifying fillna for every column using dictionary format.
# forward fill with na and backward fill
# axis option in fillna
```

```python
# interpolate method, time option
df3 = df.interpolate(method="time")
df3
```

```python
# inserting new rows according to date using date_range() and DateTimeIndex() and reindex
dt = pd.date_range("01-01-2019", "12-01-2019")
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
df
```

```python
# handling  wrong data instead of missing data
# replace method - providing single value and list of values
# using dictionary to provide the columns to replace
# using dictionary to replace different entries with different values.
df4=df.replace({
    'temperature':-1,
    'windspeed':0,
    "event":"-"
}, np.NaN)
df4
```
