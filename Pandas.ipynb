```python
import pandas as pd
import numpy as np
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    print(ret)
    ret[n:] = ret[n:] - ret[:-n]
    print(ret[n:])
    return ret[n - 1:] / n
```

```python
a=np.arange(10)
print(moving_average(a, n=4))
```
